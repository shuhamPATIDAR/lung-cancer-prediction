As we can seen, values of attributes in dataset is of numeric type,and YES and NO in case of LUNG_CANCER attribute. So at first, we are using the label encoder to convert this values in 0 and 1. Where 0 denotes 'NO' and 1 denoted 'Yes'

```python
#importing preprocessing from sklearn.
#Method fit_transform convert the values.

from sklearn import preprocessing

le=preprocessing.LabelEncoder()
df['GENDER']=le.fit_transform(df['GENDER'])
df['SMOKING']=le.fit_transform(df['SMOKING'])
df['YELLOW_FINGERS']=le.fit_transform(df['YELLOW_FINGERS'])
df['ANXIETY']=le.fit_transform(df['ANXIETY'])
df['PEER_PRESSURE']=le.fit_transform(df['PEER_PRESSURE'])
df['CHRONIC DISEASE']=le.fit_transform(df['CHRONIC DISEASE'])
df['FATIGUE ']=le.fit_transform(df['FATIGUE '])
df['ALLERGY ']=le.fit_transform(df['ALLERGY '])
df['WHEEZING']=le.fit_transform(df['WHEEZING'])
df['ALCOHOL CONSUMING']=le.fit_transform(df['ALCOHOL CONSUMING'])
df['COUGHING']=le.fit_transform(df['COUGHING'])
df['SHORTNESS OF BREATH']=le.fit_transform(df['SHORTNESS OF BREATH'])
df['SWALLOWING DIFFICULTY']=le.fit_transform(df['SWALLOWING DIFFICULTY'])
df['CHEST PAIN']=le.fit_transform(df['CHEST PAIN'])
df['LUNG_CANCER']=le.fit_transform(df['LUNG_CANCER'])
```

```python
#Splitting Independent and Dependent Features
X=df_new.drop('LUNG_CANCER',axis=1)
Y=df_new['LUNG_CANCER']
X
```

```python
#Fixing the imbalance dataset using ADASYN technique
from imblearn.over_sampling import ADASYN
adasyn =ADASYN(random_state=42)
X,Y=adasyn.fit_resample(X,Y)
```

```python
len(X)
```

## Logistic Regression

```python
#Logestic Regression

#splitting the dataset for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.25,random_state=0)
#Here random_state =0 state that the split will be same every time you run the code
```

```python
#Fitting the training data to the model

from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train,Y_train)
```

```python
#predinting the result using the test data
Y_lr_predict=lr_model.predict(X_test)
Y_lr_predict
```

```python
#MOdel Accuracy

from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(Y_test,Y_lr_predict)
print(lr_cr)
```

THIS MODEL IS ALMOST 95% accurate

## Decision Tree

```python
#Fitting the data to the model

from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_model.fit(X_train,Y_train)
```

```python
#predicting the result using the testing data

Y_dt_predict=dt_model.predict(X_test)
print(Y_dt_predict)
```

```python
#model accuracy

dt_cr=classification_report(Y_test, Y_dt_predict)
print(dt_cr)
```

Decision tree Accuracy is almost 95%

## k Nearest Nieghbour

```python
#Fitting the K-NN classifier to the training dataset

from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
knn_model.fit(X_train,Y_train)

#n_neighbors parameter decide specific number of nieghbour to be consider
#metric ='minkowski' discribe the type of distance matric used.The Minkowski distance is a generalization of both Euclidean and Manhattan distances.
#p=2: This parameter is specific to the Minkowski distance. When p=2, it calculates the Euclidean distance, which is the straight-line distance between two points in space.
```

```python
#Predicting the result using testing data
Y_knn_perdict=knn_model.predict(X_test)
Y_knn_perdict
```

```python
#Model Accuracy
knn_cr=classification_report(Y_test,Y_knn_perdict)
print(knn_cr)
```

Accuracy of model is almost 95%

## K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

k=10
kf=KFold(n_splits=k, shuffle=True, random_state=42)

#Logistic regression model
lr_model_scores=cross_val_score(lr_model,X,Y,cv=kf)

# Decision tree model
dt_model_scores =cross_val_score(dt_model,X,Y,cv=kf)

#Gaussian naive bayes model
gnb_model_scores = cross_val_score(gnb_model,X,Y,cv=kf)

# KNN model
knn_model_scores = cross_val_score(knn_model,X, Y, cv=kf)

# Multinomial naive bayes model
mnb_model_scores = cross_val_score(mnb_model,X, Y, cv=kf)

# Support Vector Classifier model
svc_model_scores = cross_val_score(svc_model,X, Y, cv=kf)

# Random forest model
rf_model_scores = cross_val_score(rf_model,X, Y, cv=kf)


# Multi-layer perceptron model
mlp_model_scores = cross_val_score(mlp_model,X, Y, cv=kf)




print("Logistic regression models' average accuracy:", np.mean(lr_model_scores))
print("Decision tree model's average acccuracy :", np.mean(dt_model_scores))

print("KNN models' average accuracy:", np.mean(knn_model_scores))
print("Gaussian naive bayes models' average accuracy:", np.mean(gnb_model_scores))
print("Multinomial naive bayes models' average accuracy:", np.mean(mnb_model_scores))
print("Support Vector Classifier models' average accuracy:", np.mean(svc_model_scores))
print("Random forest models' average accuracy:", np.mean(rf_model_scores))
print("Multi-layer perceptron models' average accuracy:", np.mean(mlp_model_scores))
```
